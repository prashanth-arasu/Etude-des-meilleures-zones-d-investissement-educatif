<h3><center><em>I. 2. Préanalyse des données</em></center></h3>

```python
# Qualité du jeu des données

print("D'abord, nous allons analyser la taille et la qualité des données de chaque dataframe.")
print()
nom_dfs = ["country", "cseries", "data", "foot", "series"]
i = 0
dfs = [country, cseries, data, foot, series]
df_not_null = [None] * 5
df_null = [None] * 5
for df in dfs:
    print("La dataframe", nom_dfs[i], end = "")
    print(" consiste de:")
    print("Lignes:", len(df))
    print("Colonnes:", len(df.columns))
    print("Elements: ", df.size)
    a = df.isna().sum().sum()*100/df.size
    print("Données manquantes: " + str(a) + "%")
    b = df.duplicated().sum()
    print("Données dupliquées: ", b)
    print("")
    df_not_null[i] = df.size - df.isna().sum().sum()
    df_null[i] = df.isna().sum().sum()
    i+=1
```

```python
# Plot 

width = 0.3

fig = plt.figure()

X=np.arange(5)

ax = fig.add_axes([0,0,1,1])
plt.xticks([r + width for r in range(len(nom_dfs))], nom_dfs)
ax.bar(X + 0.00, df_not_null, width, color='b')
ax.bar(X + 0.3, df_null, width, color='r')
ax.set_yscale('log')
plt.xlabel('Dataframes')
plt.ylabel('Elements (log)')
ax.legend(labels=['Valid Elements', 'Null Elements'])
plt.ylim(1000, 100000000)
ax.set_title("Nombres d'elements par dataframe")
plt.show()
```

```python
# Calcul du nombre des données manquantes par pay 

column_names = ["Country Name", "Null Data in %"]
null_pays = pd.DataFrame(columns = column_names)

nom_pays = pd.unique(data_pays['Country Name'])

null_data_pays_percent = [None] * 217
for i in range (0,217):
    temp = data_pays.iloc[0*i:3665*(i+1),:]
    null_data_pays_percent[i]=temp.isnull().sum().sum()*100/temp.size

null_pays["Country Name"] = nom_pays
null_pays["Null Data in %"] = null_data_pays_percent

# Calcul du nombre des données manquantes par zone

column_names_1 = ["Zone Name", "Null Data in %"]
null_zones = pd.DataFrame(columns = column_names_1)

nom_zones = pd.unique(data_zones['Country Name'])

null_data_zones_percent = [None] * 25
for i in range (0,25):
    temp = data_zones.iloc[0*i:3665*(i+1),:]
    null_data_zones_percent[i]=temp.isnull().sum().sum()*100/temp.size

null_zones["Zone Name"] = nom_zones
null_zones["Null Data in %"] = null_data_zones_percent
```

```python
# Analyse statistique des données manquantes

nulls = [null_pays, null_zones]
nom_nulls = ["pay", "zone"]
j = 0
for null in nulls:
    print("Statistiques des données par " + nom_nulls[j] + ":")
    print("")
    print("Minimum:", min(null["Null Data in %"]))
    print("Maximum:", max(null["Null Data in %"]))
    print("Moyenne:", st.mean(null["Null Data in %"]))
    print("Écart type:", st.pstdev(null["Null Data in %"]))
    print("")
    j+=1
```

<h3><center><em>I. 4. Système de scoring</em></center></h3>

Nous proposons un système de scoring pour évaluer le potentiel et l'évolution du potentiel pour chaque pay et zone. Ce score est donné par le cumul du score de chaque critère.

<h4>Score du potentiel:</h4>

Total Score = Population Score + Formation Score + Ressources Score

<h4>Score de l'évolution du potentiel:</h4>

Evolution Score = Population Evl Score + Formation Evl Score

Le scoring du chaque critère est obtenu des scores de chaque enquête. 

<h4>1) La population de la clientèle (0-25 ans)</h4>

<h5>Potentiel:</h5>

Population Score = 0.6 * (Population, ages 0-14, total) + 0.4*(Population, ages 15-24, total)

<h5>Évolution du potentiel:</h5>

Population Evl Score = (Population growth (annual %))

<h4>2) La population des éleves en formation</h4>

<h5>Potentiel:</h5>

Formation Score =   0.6 * (Barro-Lee: Average years of total schooling, age 15+, total) +
                    0.4 * (Barro-Lee: Average years of total schooling, age 25+, total)
                    

<h5>Évolution du potentiel:</h5>

Formation Evl Score =

0.8 * (Wittgenstein Projection: Mean years of schooling. Age 0-19. Total) + 0.2 * (Wittgenstein Projection: Mean years of schooling. Age 20-24. Total)

<h4>3) L'accès aux ressources informatiques</h4>

<h5>Potentiel:</h5>

Formation Score =   0.4 * (Internet users (per 100 people)) + 0.3 * (GDP per capita (current US)) + 
                    0.3 * (Personal computers (per 100 people))

<h3><center><em>II. 2. Évaluation de l'évolution du potentiel de clients</em></center></h3>
Dans cette section, nous évaluerons le potentiel des pays et des zones sélectionnés suite à notre première analyse. D'abord, nous définissons des nouveaux dataframes pour cette analyse. 

```python
fort_potentiel_pays=['USA','CHN','IND','BRA','NOR','GBR','DNK','BHR','KOR','HKG','SWE','CHE',
                     'SGP','DEU','FIN','NLD','NZL','CAN','FRA','AUT','IRL']
fort_potentiel_zones=['EAS','ECS','NAC','EUU', 'SAS']

data_pays_sel=data_pays[data_pays["Country Code"].isin(fort_potentiel_pays)]
data_pays_sel=data_pays_sel.reset_index(drop=True)

data_zones_sel=data_zones[data_zones["Country Code"].isin(fort_potentiel_zones)]
data_zones_sel=data_zones_sel.reset_index(drop=True)

score_list_zones_sel=score_list_zones[score_list_zones["Country Code"].isin(fort_potentiel_zones)]
score_list_zones_sel=score_list_zones_sel.reset_index(drop = True)
```

```python
pop_growth = data_pays_sel[data_pays_sel['Indicator Code'] == 'SP.POP.GROW']
pop_growth = pop_growth.reset_index(drop = True)
pop_growth = pop_growth[['Country Name','2016']]
pop_growth.columns = ["Country Name", "Values"]

proj_edu_pri = data_pays_sel[data_pays_sel['Indicator Code']=='PRJ.POP.ALL.1.MF']
proj_edu_pri = proj_edu_pri.reset_index(drop=True)
proj_edu_pri = proj_edu_pri[['Country Name','2025','2030']]

proj_edu_sec = data_pays_sel[data_pays_sel['Indicator Code']=='PRJ.POP.ALL.3.MF']
proj_edu_sec = proj_edu_sec.reset_index(drop=True)
proj_edu_sec = proj_edu_sec[['Country Name','2025','2030']]

proj_edu_postsec = data_pays_sel[data_pays_sel['Indicator Code']=='PRJ.POP.ALL.4.MF']
proj_edu_postsec = proj_edu_postsec.reset_index(drop=True)
proj_edu_postsec = proj_edu_postsec[['Country Name','2025','2030']]

pop_growth_zones = data_zones_sel[data_zones_sel['Indicator Code'] == 'SP.POP.GROW']
pop_growth_zones = pop_growth_zones.reset_index(drop = True)
pop_growth_zones = pop_growth_zones[['Country Name','2013']]
pop_growth_zones.columns = ["Country Name", "Values"]
```

Nous attribuons des score pour les croissances de population par pays et par zones.

```python
df_pop_growths = [pop_growth, pop_growth_zones]
for df_pop_growth in df_pop_growths:
    mean = st.mean(df_pop_growth['Values'])
    stdev = st.pstdev(df_pop_growth['Values'])
    limit_a = mean + (1.5*stdev)
    limit_b = mean + (0.5*stdev)
    limit_c = mean - (0.5*stdev)
    limit_d = mean - (1.5*stdev)
    df_pop_growth.loc[df_pop_growth['Values'] < limit_d, ['Score']] = 1
    df_pop_growth.loc[(df_pop_growth['Values'] > limit_d) & (df_pop_growth['Values'] < limit_c), ['Score']] = 2
    df_pop_growth.loc[(df_pop_growth['Values'] > limit_c) & (df_pop_growth['Values'] < limit_b), ['Score']] = 3
    df_pop_growth.loc[(df_pop_growth['Values'] > limit_b) & (df_pop_growth['Values'] < limit_a), ['Score']] = 4
    df_pop_growth.loc[df_pop_growth['Values'] > limit_a, ['Score']] = 5
```

Pour l'évaluation de l'évolution du potentiel de clients, nous définissons un nouveau système de score. Pour cela, nous utiliserons des valeurs ajustés des nombres d'éleves pour l'année 2020 et 2025. 

```python
column_names_4 = ["Country Name", "Country Code","Pop_Growth Score", "Adjusted 2025", "Adjusted 2030", "Score_2025", "Score_2030"]
pot_evl_score_pays = pd.DataFrame(columns = column_names_4)
```

```python
pot_evl_score_pays['Country Name'] = score_list_sel["Country Name"]
pot_evl_score_pays['Country Code'] = score_list_sel["Country Code"]
pot_evl_score_pays['Pop_Growth Score'] = pop_growth["Score"]
pot_evl_score_pays['Adjusted 2025'] = (0.4 * proj_edu_pri["2025"] + 0.4 * proj_edu_sec["2025"] + 0.2 *proj_edu_postsec["2025"])
pot_evl_score_pays['Adjusted 2030'] = (0.4 * proj_edu_pri["2030"] + 0.4 * proj_edu_sec["2030"] + 0.2 *proj_edu_postsec["2030"])

mean_2025 = st.mean(pot_evl_score_pays['Adjusted 2025'])
stdev_2025 = st.pstdev(pot_evl_score_pays['Adjusted 2025'])
limit_2025_a = mean_2025 + (1.5*stdev_2025)
limit_2025_b = mean_2025 + (0.5*stdev_2025)
limit_2025_c = mean_2025 - (0.5*stdev_2025)
limit_2025_d = mean_2025 - (1.5*stdev_2025)

pot_evl_score_pays.loc[pot_evl_score_pays['Adjusted 2025']< limit_2025_d, 'Score_2025'] = 1
pot_evl_score_pays.loc[(pot_evl_score_pays['Adjusted 2025']> limit_2025_d) & (pot_evl_score_pays['Adjusted 2025']< limit_2025_c), 'Score_2025'] = 2
pot_evl_score_pays.loc[(pot_evl_score_pays['Adjusted 2025']> limit_2025_c) & (pot_evl_score_pays['Adjusted 2025']< limit_2025_b), 'Score_2025'] = 3
pot_evl_score_pays.loc[(pot_evl_score_pays['Adjusted 2025']> limit_2025_b) & (pot_evl_score_pays['Adjusted 2025']< limit_2025_a), 'Score_2025'] = 4
pot_evl_score_pays.loc[pot_evl_score_pays['Adjusted 2025']> limit_2025_a, 'Score_2025'] = 5

mean_2030 = st.mean(pot_evl_score_pays['Adjusted 2030'])
stdev_2030 = st.pstdev(pot_evl_score_pays['Adjusted 2030'])
limit_2030_a = mean_2030 + (1.5*stdev_2030)
limit_2030_b = mean_2030 + (0.5*stdev_2030)
limit_2030_c = mean_2030 - (0.5*stdev_2030)
limit_2030_d = mean_2030 - (1.5*stdev_2030)
                                          
pot_evl_score_pays.loc[pot_evl_score_pays['Adjusted 2030']< limit_2030_d, 'Score_2030'] = 1
pot_evl_score_pays.loc[(pot_evl_score_pays['Adjusted 2030']> limit_2030_d) & (pot_evl_score_pays['Adjusted 2030']< limit_2030_c), 'Score_2030'] = 2
pot_evl_score_pays.loc[(pot_evl_score_pays['Adjusted 2030']> limit_2030_c) & (pot_evl_score_pays['Adjusted 2030']< limit_2030_b), 'Score_2030'] = 3
pot_evl_score_pays.loc[(pot_evl_score_pays['Adjusted 2030']> limit_2030_b) & (pot_evl_score_pays['Adjusted 2030']< limit_2030_a), 'Score_2030'] = 4
pot_evl_score_pays.loc[pot_evl_score_pays['Adjusted 2030']> limit_2030_a, 'Score_2030'] = 5

pot_evl_score_pays['Evolution Score'] = (0.25* pot_evl_score_pays['Pop_Growth Score'] + 0.4* pot_evl_score_pays['Score_2025']+ 0.35* pot_evl_score_pays['Score_2030'])
pot_evl_score_pays['Score du potentiel'] = score_list_sel['Score du potentiel']
pot_evl_score_pays['Total Score'] = pot_evl_score_pays['Evolution Score']+ score_list_sel['Score du potentiel']
pot_evl_score_pays
```
